```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import plotly.express as px


eventos=pd.read_csv("/content/drive/MyDrive/mumu.csv",header=0,index_col=0,delimiter=',',dtype=np.float64)
eventos.head()
```

```python
p=np.sqrt(eventos['px']**2+eventos['py']**2+eventos['pz']**2)
pz=eventos['pz']

x=pz/p

def eta(x):
  theta=np.arccos(x)
  eta=-np.log(np.tan(theta/2))
  return eta

eta=pd.DataFrame(eta(x))

fig_0=px.scatter(x=theta,y=eta)
fig_0.update_layout(
    title=r"$\theta  \ vs \ \eta$",
    xaxis_title=r"$\theta \text{[rad]} $",
    yaxis_title=r"$ \eta $")
fig_0.show()

```
