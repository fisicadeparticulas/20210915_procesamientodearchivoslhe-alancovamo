# 2. Construccion de la funcion correcta para $\phi$

Usando ciclos de iteracion y condicionales, definimos una función para obtener el ángulo correcto. Usamos la convención de tomar ángulos referidos a $x$ de $0$ a $\pi$ y de $0$ a $-\pi$.

```python
#Definimos la funcion para obtener el valor adecuado del arcotangente

def arctan(py,px):
    phi=[]
    for i,j in zip(py,px):
        if (j>=0):
            if (i>=0):
                phi.append(np.arctan(i/j))
            else:
                phi.append(np.arctan(i/j))
        else:
            if (i>=0):
                phi.append(np.pi+np.arctan(i/j))
            else:
                phi.append(-np.pi+np.arctan(i/j))
    return np.array(phi)


x=np.array([1,-1,-1,1])
y=np.array([1,1,-1,-1])
print(arctan(y,x)*180/np.pi)

#Realizamos una pequeña comprobacion de nuestra función, vemos que mapea correctamente los vectores
#(1,1),(-1,1),(-1,-1) y (1,-1) a los angulos correspondientes.
```

```python
#Creamos las figuras
phi_1=arctan(eventos['py*'],eventos['px*'])
phi_2=arctan(eventos['py'],eventos['px'])

fig_0=px.histogram(phi_1,x=phi_1[:],nbins=14)
fig_0.update_layout(
    title=r"$ \text{ Distribucion de }  \phi_1$",
    xaxis_title="Radianes",
    yaxis_title="Eventos")
fig_0.show()
fig_1=px.histogram(phi_2,x=phi_2[:],nbins=14)
fig_1.update_layout(
    title=r"$ \text{ Distribucion de }  \phi_2$",
    xaxis_title="Radianes",
    yaxis_title="Eventos")
fig_1.show()
```

```python
phi_1=arctan(eventos['py'],eventos['px'])
phi_2=arctan(eventos['py*'],eventos['px*'])

dphi=np.abs(phi_2-phi_1)

fig=px.line(dphi,x=dphi[:],range_x=[3.1,3.2])
fig.update_layout(
    title=r"$ \text{ Distribucion de }  d\phi$",
    xaxis_title="Radianes",
    yaxis_title="Eventos")
fig.show()
```

# Todos los eventos devuelven un ángulo $d\phi=\pi$
